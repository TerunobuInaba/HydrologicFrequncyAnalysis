```python
# importing necessary packages
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pylab import rcParams
from scipy.stats import norm
from scipy.stats import lognorm
import numpy as np

# reading data
df = pd.read_csv("ppt_fukuoka.csv")

# setting figure 
rcParams["figure.figsize"] = 5, 3
```

```python
# plotting a histgram for annual precipitation
annual = df["Annual PPT"]
annual
annual.hist()
plt.tight_layout()
```

```python
param = norm.fit(annual)
param
```

```python
x = np.linspace(900, 3100, 100)
pdf_fitted = norm.pdf(x, loc=param[0], scale=param[1])
plt.plot(x, pdf_fitted, 'r-')
annual.hist(density=True, alpha=0.8)
plt.show()
```

```python
dailyMax = df["Max Daily PPT"]
dailyMax.hist()
plt.tight_layout()
```

```python
param = lognorm.fit(dailyMax)
param
```

```python
x = np.linspace(40, 310, 100)
pdf_fitted = lognorm.pdf(x, param[0], loc=param[1], scale=param[2])
plt.plot(x, pdf_fitted, 'r-')
dailyMax.hist(density=True, alpha=0.8)
plt.show()
```
